##3) Calculate the following proportions

People with target ">50K" over total distinguished by "Race"

```python
#Series with the differents ethnic groups and their total agregations
s1_race = df['ID'].groupby(df['Race']).count()

#Application of filter >50K
mas50 = df[df['Target'] == '>50K']
s2_race = mas50['ID'].groupby(mas50['Race']).count()

#Calculation of the proportion
race_per = s2_race.values/s1_race.values
race_proportion = pd.Series(data=race_per, index=s1_race.index)
```

```python
race_proportion
```
