```python
import pandas as pd
```

```python
olympics_file = pd.read_csv('Summer-Olympic-medals-1976-to-2008.csv', encoding = "ISO-8859-1")
```

```python
olympics_file
```

```python
print(olympics_file.dtypes)
```

```python
olympics_file['Year'] = (olympics_file['Year'].fillna(0.0)).astype(int)
```

```python
df_olympics_file = olympics_file.set_index('Year')
df_olympics_file
```

```python
df_olympics_file[['Lastname','Firstname']] = df_olympics_file['Athlete'].str.split(',', 1, expand=True)
df_olympics_file.drop('Athlete', inplace=True, axis=1)
df_olympics_file
```

```python
df_olympics_file.describe()
```

```python
country_medals = df_olympics_file.groupby(df_olympics_file['Country']).Country.count()
country_medals.sort_values(ascending = False)
```

```python
df_men = df_olympics_file[df_olympics_file.Gender == 'Men']
men_medals = df_men.groupby(df_men.index).Gender.count()
men_medals
```
